Verificar quais as features mais significativas para esta análise, através de t-test:

```python
# Lista vazia para armazenar os p-values
pvalues = []

# Separar os dados com base nos valores de Y
data_y0 = features_final[features_final['Y'] == 0]
data_y1 = features_final[features_final['Y'] == 1]

# Realizar o teste t para cada coluna
for col in features_final.columns[:-1]:
    t_stat, p_value = ttest_ind(data_y0[col], data_y1[col])
    print(f"{col}: t-statistic = {t_stat:.4f}, p-value = {p_value:.5f}")

# Armazena o p-value na lista
    pvalues.append(p_value)

# Cria uma série com os p-values e os nomes das colunas
pvalues = pd.Series(pvalues, index=features_final.columns[:-1]).sort_values(ascending=True)

# Seleciona as 5 colunas com os menores p-values
colunas_selecionadas = pvalues.nsmallest(5).index.tolist()
```

```python
colunas_selecionadas
```

Com base nos ttest realizados, consegue-se perceber que as features mais significativas para esta análise, ou seja, as que apresentam um p-value menor,são as 'MolLogP', 'BCUT2D_MWHI', 'BCUT2D_MRHI', 'NOCount' e 'MolMR'. 

MolLogP: corresponde às medidas de log P octanol-água dos compostos químicos a analisar, que indica a hidrofobicidade (solubilidade em água) das moléculas.

BCUT2D_MRHI: representa a massa ponderada hidrofóbica dos anéis aromáticos das moléculas. 

NOCount: número de grupos nitro (-NO2) presentes numa molécula.

BCUT2D_MWHI: representa a massa ponderada hidrofóbica das ligações entre átomos de carbono das moléculas. 

MolMR: corresponde à medida da capacidade das moléculas interagirem com outras moléculas num ambiente químico.

```python
MolLogP = features_final['MolLogP']
BCUT2D_MWHI = features_final['BCUT2D_MWHI']
BCUT2D_MRHI = features_final['BCUT2D_MRHI']
NOCount = features_final['NOCount']
MolMR = features_final['MolMR']
```

Testes estatísticos (t-test) para as 5 features mais significativas:

H0: Não existem diferenças significativas entre o composto químico e a inibição de CYP2C19.

H1: Existem diferenças significativas entre o composto químico e a inibição de CYP2C19.

```python
#Feature MolLogP
med_inhib = features_final[features_final["Y"] == 1]["MolLogP"]
med_not_inhib = features_final[features_final["Y"] == 0]["MolLogP"]
t_stat, p_val = ttest_ind(med_inhib, med_not_inhib)

# consultar p-value
if p_val < 0.05:
    print('p_value:', p_val)
    print('Rejeita-se H0, pelo que, para um grau de confiança de 95%, conclui-se que existem diferenças significativas entre MolLogP e a inibição de CYP2C19.')
else:
    print('p_value:', p_val)
    print('Não se rejeita H0, pelo que, para um grau de confiança de 95%, não se pode afirmar que existem diferenças significativas entre MolLogP e a inibição de CYP2C19.')
```

Grupos próximos no dendrograma podem representar clusters de amostras com propriedades similares. Estas features podem partilhar características.
Os clusters são mesclados em alturas diferentes, o que indica a proximidade ou distância entre eles. Podemos observar a formação de grupos maiores e grupos menores conforme os clusters são mesclados, ou seja, a feature 'MaxAbsEStateIndex' apresenta maior distância em relação às restantes. 

```python
def plot_dendrogram(model, **kwargs):
    # Create linkage matrix and then plot the dendrogram

    # create the counts of samples under each node
    counts = np.zeros(model.children_.shape[0])
    n_samples = len(model.labels_)
    for i, merge in enumerate(model.children_):
        current_count = 0
        for child_idx in merge:
            if child_idx < n_samples:
                current_count += 1  # leaf node
            else:
                current_count += counts[child_idx - n_samples]
        counts[i] = current_count

    linkage_matrix = np.column_stack(
        [model.children_, model.distances_, counts]
    ).astype(float)

    # Plot the corresponding dendrogram
    dendrogram(linkage_matrix, **kwargs)



X = feat_sel_pca_scale

# setting distance_threshold=0 ensures we compute the full tree.
model = AgglomerativeClustering(distance_threshold=0, n_clusters=None)

model = model.fit(X)
plt.title("Hierarchical Clustering Dendrogram")
# plot the top three levels of the dendrogram
plot_dendrogram(model, truncate_mode="level", p=2)
plt.xlabel("Number of points in node.")
plt.show()
```

Em cima, é apresentado um dendrograma com um único cluster hierárquico que contém todas as amostras. 

```python
# k-means clustering
kmeans = KMeans(n_clusters=2, max_iter=1000)
kmeans.fit(features_sig_scale)
labels_k = kmeans.labels_
centroids = kmeans.cluster_centers_

pd.crosstab(labels_k,features_sig_y['Y'], rownames=['clusters'] )
```

##### K-Nearest Neighbors
O algoritmo K-Nearest Neighbors (KNN) é um método de machine learning utilizado para classificação e regressão. O KNN determina a classe de um ponto de dados de teste encontrando os k vizinhos mais próximos desse ponto e atribuindo a classe mais comum entre esses vizinhos ao ponto de teste. O valor de k, que é definido previamente, determina quantos vizinhos são considerados na classificação.

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix
import math
from math import sqrt
from sklearn.metrics import accuracy_score, precision_score, recall_score, roc_auc_score, mean_squared_error, f1_score

knn = KNeighborsClassifier()
knn.fit(train_in, train_out)
preds = knn.predict(test_in)

print("Valores previstos: ", preds)
print("Valores reais: " , test_out)
print("Perc. acertos: ", (preds == test_out).sum()/len(preds) )


print("Matriz de confusão:")
confusion_matrix(test_out, preds, labels = [0,1])

accuracy = accuracy_score(test_out, preds)
print("Accuracy:", accuracy)

precision = precision_score(test_out, preds)
print("Precision:", precision)

recall = recall_score(test_out, preds)
print("Recall:", recall)

f1 = f1_score(test_out, preds)
print("F1-score:", f1)

auc_roc = roc_auc_score(test_out, preds)
print("AUC-ROC:", auc_roc)

mean_error = mean_squared_error(test_out, preds)  
root_mean = sqrt(mean_error)
mean_error = mean_squared_error(test_out, preds)
root_mean = math.sqrt(mean_error)
root_mean
print("Raiz quadrada do erro:", root_mean)

print()
```

```python
print("Matriz de confusão:")
confusion_matrix(test_out, preds, labels = [0,1])
```

A matriz de confusão indica que o modelo KNN classificou corretamente 394 instâncias da classe "0" e 449 instâncias da classe "1". No entanto, clasificou erroneamente 180 instâncias da classe "0" como "1" e 212 instâncias da classe "1" como "0". Relativamente às métricas de desempenho, a accuracy obtida para o modelo KNN foi de aproxidamente 68%, indicando um bom nível de precisão do modelo.

#### Deep Learning - Deep neural networks

```python
from tensorflow.keras import models, layers
#Definir a estrutura do modelo:
hidden = 8

model = models.Sequential()
model.add(layers.Dense(hidden, activation='relu', input_shape=(5,)))
model.add(layers.Dense(hidden, activation='relu'))
model.add(layers.Dense(1, activation='sigmoid'))
```

```python
#Treinar o modelo:
model.compile(optimizer='rmsprop',
              loss='binary_crossentropy',
              metrics=['acc'],
              run_eagerly=True)

val_in = train_in[:1000]
partial_train_in = train_in[1000:]
val_out = train_out[:1000]
partial_train_out = train_out[1000:]


history = model.fit(partial_train_in,
                    partial_train_out,
                    epochs=50,
                    batch_size=32,
                    validation_data=(val_in, val_out))
```

```python
history_dict = history.history
acc_values = history_dict['acc'] 
val_acc_values = history_dict['val_acc']
epochs = range(1, len(acc_values) + 1)

import matplotlib.pyplot as plt
plt.plot(epochs, acc_values, 'bo', label='Training acc')
plt.plot(epochs, val_acc_values, 'b', label='Validation acc')
plt.title('Training and validation accuracy')
plt.xlabel('Epochs')
plt.ylabel('Acc')
plt.legend()
plt.show()
```

Através da análise gráfica, os valores de precisão (accuracy) do conjunto de treino e do conjunto de validação são próximos. Isto indica que o modelo está a ser bem treinado, o que indica um bom equilíbrio entre os dados.

```python
loss_values = history_dict['loss'] 
val_loss_values = history_dict['val_loss']

plt.plot(epochs, loss_values, 'bo', label='Training loss')
plt.plot(epochs, val_loss_values, 'b', label='Validation loss')
plt.title('Training and validation loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

Através da análise gráfica, pode-se monitorizar a perda (loss) nos conjuntos de treino e validação. A perda é uma medida do quão bem o modelo está a ser ajustado aos dados durante o treino. A perda de treino assim como a perda de validação diminuem à medida que o modelo é treinado, apresentando valores baixos, indicando que o modelo está a fazer previsões mais precisas. Não se verifica uma grande diferença entre as perdas de treino e validação.

Portanto, é importante monitorizar tanto a precisão (accuracy) quanto a perda (loss) do treino e validação para avaliar o desempenho do modelo e fazer ajustes, se necessário.

```python
results_deep = model.evaluate(test_in, test_out, verbose = 0)
results_deep
```

Nesse caso, a perda obtida foi de aproximadamente 59% e a accuracy alcançada foi de aproximadamente 69%.
A accuracy indica a percentagem de amostras de treino que foram classificadas corretamente pelo modelo.
